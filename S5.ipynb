```python
import matplotlib.pyplot as plt
from utils import get_data_loader, train, test, train_losses, test_losses, train_acc, test_acc
from model import get_train_parameters
```

```python
num_epochs = 20
batch_size = 512
```

```python
train_loader, test_loader = get_data_loader(batch_size)
```

```python
batch_data, batch_label = next(iter(train_loader)) 
fig = plt.figure()
for i in range(12):
  plt.subplot(3,4,i+1)
  plt.tight_layout()
  plt.imshow(batch_data[i].squeeze(0), cmap='gray')
  plt.title(batch_label[i].item())
  plt.xticks([])
  plt.yticks([])
```

```python
model, device, optimizer, scheduler, criterion = get_train_parameters()
```

```python
#Iterate over each epoch and calculate test accuracy after each epoch
for epoch in range(1, num_epochs+1):
  print(f'Epoch {epoch}')
  train(model, device, train_loader, optimizer, criterion)
  test(model, device, test_loader, criterion)
  scheduler.step()
```

```python
fig, axs = plt.subplots(2,2,figsize=(15,10))
axs[0, 0].plot(train_losses)
axs[0, 0].set_title("Training Loss")
axs[1, 0].plot(train_acc)
axs[1, 0].set_title("Training Accuracy")
axs[0, 1].plot(test_losses)
axs[0, 1].set_title("Test Loss")
axs[1, 1].plot(test_acc)
axs[1, 1].set_title("Test Accuracy")
```
